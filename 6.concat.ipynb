```python
import pandas as pd

india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
import pandas as pd

us_weather = pd.DataFrame({
    "city": ["newyork","chicago","vegas"],
    "temperature": [22,35,20],
    "humidity": [70, 80, 68]
})
us_weather
```

```python
pd.concat([india_weather,us_weather],ignore_index=True)
```

```python
df=pd.concat([india_weather,us_weather],keys=["INDIA","US"])
df
```

```python
df.loc["US"]
```

```python
temp_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30]})
temp_df
```

```python
hum_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "humidity": [70, 80]},index=[1,0])
hum_df
```

```python
df=pd.concat([temp_df,hum_df],ignore_index=True)
df
```

```python
df=pd.concat([temp_df,hum_df],axis=1)
df
```

```python
s=pd.Series(["humid","dry","rain"],name="event")
```

```python
df=pd.concat([temp_df,s],axis=1)
df
```
