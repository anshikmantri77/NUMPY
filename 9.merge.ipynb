```python
import pandas as pd
df=pd.DataFrame({
    "city":["newyork","chicago","chaman"],
    "temp":[21,34,24]
})
df
```

```python
df1=pd.DataFrame({
    "city":["newyork","chicago","DC","vegas"],
    "humidity":[20,23,27,29]
})
df1
```

```python
df3=pd.merge(df1,df,on="city",how="outer")
df3
```

```python
df3=pd.merge(df1,df,on="city",how="inner")
df3
```

```python
df3=pd.merge(df1,df,on="city",how="left")
df3
```

```python
df3=pd.merge(df1,df,on="city",how="right")
df3
```

```python
df3=pd.merge(df1,df,on="city",how="outer",indicator='true')
df3
```
